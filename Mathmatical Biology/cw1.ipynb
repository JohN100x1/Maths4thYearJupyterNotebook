```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
alpha = 1
eps = 0.1
w = np.linspace(eps,1/(2*np.sqrt(alpha)),N//2)
u1 = (1+np.sqrt(1-4*alpha*w**2))/(2*w)
u2 = (1-np.sqrt(1-4*alpha*w**2))/(2*w)

plt.plot(w,u1,"--",w,u2)
plt.xlabel("w")
plt.ylabel("u")
plt.legend(["u+","u-"])
plt.title("Fixed points as functions of w, alpha = {}".format(alpha))
```

```python
N = 1000
alpha = 0.5
b = 0.12
c = 2
u = np.linspace(0,8,N)
v = u/(b*(alpha+u**2))+u

plt.plot(u,v,[c,c],[0,max(v)])
plt.xlabel("u")
plt.ylabel("v")
plt.title("Nullclines alpha={}, b={}, c={}".format(alpha,b,c))
plt.annotate('f1<0, f2<0', xy=(150, 150), xycoords='figure points')
plt.annotate("", xy=(3,3.5), xytext=(3.5, 4),arrowprops=dict(arrowstyle="->"))
plt.annotate('f1<0, f2>0', xy=(65, 150), xycoords='figure points')
plt.annotate("", xy=(1,4), xytext=(1.5, 3.5),arrowprops=dict(arrowstyle="->"))
plt.annotate('f1>0, f2>0', xy=(65, 230), xycoords='figure points')
plt.annotate("", xy=(1.5,7.5), xytext=(1, 7),arrowprops=dict(arrowstyle="->"))
plt.annotate('f1>0, f2<0', xy=(150, 230), xycoords='figure points')
plt.annotate("", xy=(3.5,7), xytext=(3, 7.5),arrowprops=dict(arrowstyle="->"))
plt.legend(["f1=0", "f2=0"],loc="lower right")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

alpha = 0.1
b = 0.1
c = 0.5

# Nullclines
u = np.linspace(0,c,1000)
f1 = u/(alpha+u**2)
f2 = c/(alpha+u**2)
plt.xlabel("u")
plt.ylabel("w")


#plt.plot(u,f2)

def du_dt(u, t):
    return [u[1]*(alpha + u[0]**2) - u[0], b*(c-u[1]*(alpha+u[0]**2))]

ts = np.linspace(0, 200, 100000)
u0 = [0, 0]
us = odeint(du_dt, u0, ts)
pop1 = us[:,0]
pop2 = us[:,1]
print(1/(2*np.sqrt(alpha)))
print(max(pop2))
plt.plot(u,f1)
plt.plot([np.sqrt(alpha),np.sqrt(alpha)],[0,max(pop2)])
plt.plot(pop1,pop2);
print(ts[list(us[:,1]).index(max(pop2))])
```

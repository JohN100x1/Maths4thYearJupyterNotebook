## 1. Spruce budworm model with spatial dynamics

\begin{enumerate}

\item Assuming a travelling wave solution $U(\xi)$ where $\xi = x - ct$, the PDE becomes 

\begin{equation*}
-c \frac{dU}{d\xi} = f(U) + \frac{d^2 U}{d\xi ^2}
\end{equation*}

Defining $V = dU/d\xi$, we have the first-order system 

\begin{equation*}
\frac{dU}{d\xi} = V \\
\frac{dV}{d\xi} = -f(U) - cV
\end{equation*}

\item The fixed points are $(0,0),\, (u_1,0),\, (u_2,0),\, \text{and } (u_3,0)$, where $u_i$ satisfy $f(u_i) = 0$.
The Jacobian of the system is 

\begin{equation*}
J(U,V)
\begin{bmatrix}
0 & 1 \\
-f'(U) & -c
\end{bmatrix}
\end{equation*}

where $f'(U) = df/dU$. The eigenvalues are then 

\begin{equation*}
\lambda_{\pm} = \frac{-c \pm \sqrt{c^2 - 4\frac{df}{dU}}}{2}.
\end{equation*}

Based on the stability of the fixed points of the system $\frac{du}{dt} = f(u)$, we know $f'(0)>0,\, f'(u_1)<0,\,f'(u_2)>0\,, \text{and } f'(u_3)<0 .$

Thus, from the expression for the eigenvalues, and taking $c>0$,

\begin{equation*}
(0,0) = 
\begin{cases}
      \text{stable node,}  & c^2 > 4f'(0)\\
       \text{stable spiral,}  & c^2 < 4f'(0)
    \end{cases} \\
(u_2,0) = 
\begin{cases}
      \text{stable node,}  & c^2 > 4f'(u_2)\\
       \text{stable spiral,}  & c^2 < 4f'(u_2)
    \end{cases} \\
(u_1,0) = \text{saddle, } \\
(u_3,0) = \text{saddle }
\end{equation*}

For $c<0$, $(0,0)$ and $(u_2,0)$ are unstable nodes or spirals for the same conditions described above when $c>0$. 

\item Since we cannot have $U<0$, $(0,0)$ cannot be a spiral. Thus, $c^2 > 4f'(0)$.

\item Multiplying the second-order equation by $dU/d\xi$ and integrating, we have

\begin{equation*}
-c \int_{-\infty}^{\infty} \left(\frac{dU}{d\xi}\right)^2 d\xi
= \int_{-\infty} ^{\infty} \frac{dU}{d\xi} f(U) d\xi + \int_{-\infty}^{\infty} \frac{dU}{d\xi} \frac{d^2U}{d\xi^2} d\xi
\end{equation*}

For the integrals on the right hand side, we have: 

\begin{equation*}
\int_{-\infty}^{\infty} \frac{dU}{d\xi}\frac{d^2U}{d\xi^2} d\xi = \frac{1}{2} \int_{-\infty}^{\infty} \frac{d}{d\xi}\left( \frac{dU}{d\xi} \right)^2 d\xi = 0
\end{equation*}

since $\frac{dU}{d\xi} \rightarrow 0$ as $\xi \rightarrow \pm \infty$.

Also, 

\begin{equation*}
\int_{-\infty}^{\infty} f(U)\frac{dU}{d\xi} d\xi = \int_{u_3}^{u_1} f(U) dU \\
= -\int_{u_1}^{u_3} f(U) dU.
\end{equation*}

Using these results, we have

\begin{equation*}
c = \frac{\int_{u_1}^{u_3} f(U) dU }{ \int_{-\infty}^{\infty} \left(\frac{dU}{d\xi}\right)^2 d\xi }
\end{equation*}

\item If we sketch $f(U)$, we see that

\begin{figure}
\includegraphics[width = \textwidth]{q1.png}
\end{figure}

We see then to ensure that $c<0$, we must have $u_2$ sufficiently close to $u_3$. This can be done by decreasing $q$, the carrying capacity for the spruce budworm. This corresponds to making their environment less hospitable. 


\end{enumerate}


## 2. Diffusive prey system

\begin{enumerate}

\item For the ODE system, we have

\begin{equation*}
\frac{du}{d\tau} = u(1-u-v) \\
\frac{dv}{d\tau} = \alpha v(u-\beta)
\end{equation*}

The fixed points are $(0,0),\, (1,0),\, \text{and } (\beta,1-\beta).$

To asses the stability of the fixed points, we consider the Jacobian:

\begin{equation*}
J(u,v) = 
\begin{bmatrix}
1-v-2u & -u \\
\alpha v & \alpha(u-\beta)
\end{bmatrix}
\end{equation*}.

At $(0,0)$, we have

\begin{equation*}
J(0,0) = 
\begin{bmatrix}
1 & 0 \\
0 & -\alpha \beta
\end{bmatrix}
\end{equation*},

giving eigenvalues

\begin{equation*}
\lambda = 1 \text{ and } \lambda = -\alpha \beta.
\end{equation*}

Thus, $(0,0)$ is a saddle.

Since $J(0,0)$ is diagonal, we see that for $\lambda = 1$, its associated eigenvector is $\boldsymbol{\xi} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$, while for $\lambda = -\alpha \beta , \boldsymbol{\xi} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$.

For fixed point $(1,0)$,
\begin{equation*}
J(1,0) = 
\begin{bmatrix}
-1 & -1 \\ 
0 & \alpha(1-\beta)
\end{bmatrix}
\end{equation*}

The eigenvalues are $\lambda = -1$ and $\lambda= \alpha (1-\beta)>0$, making $(1,0)$ a saddle. The eigenvector associated with $\lambda = -1$ is $\mathbf{\xi} = \begin{bmatrix}1 \\ 0\end{bmatrix}$.

For $(\beta, 1-\beta)$, we have 

\begin{equation*}
J(\beta, 1-\beta) = 
\begin{bmatrix}
-\beta & -\beta  \\
\alpha(1-\beta) & 0
\end{bmatrix}.
\end{equation*}

The eigenvalues are given by, 

\begin{equation*}
\lambda^2 + \beta\lambda + \beta\alpha(1-\beta) = 0,
\end{equation*}

and, therefore,

\begin{equation*}
\lambda = \frac{-\beta \pm \sqrt{\beta^2 -4\beta\alpha(1-\beta)}}{2}.
\end{equation*}

Since $4\beta\alpha(1-\beta) > 0$, we have 

\begin{equation*}
(\beta,1-\beta) = 
\begin{cases}
      \text{ stable node}& \beta^2>4\beta\alpha(1-\beta)\\
\text{ stable spiral}& \beta^2<4\beta\alpha(1-\beta)
    \end{cases}.     
\end{equation*}

A sketch of the phase diagram based on this analysis is shown below.

\begin{figure}
\includegraphics[width=\textwidth]{q2.png}
\end{figure}

\item For these conditions described in the problem, we have that $V(\xi) = 0$. i.e. there are no predators.  Thus, the equations reduce to 
\begin{equation*}
\frac{du}{d\tau} = u(1-u) + D\frac{d^2u}{dx^2}
\end{equation*}
which is the FK equation with $D \neq 1$. 

We then have for $U(\xi)$ with $\xi = x- c\tau$, that $c = 2\sqrt{D}.$

\end{enumerate}

## 3. Turing instability for the Brusselator 

\begin{enumerate}
\item We first aim to find the fixed point of the ODE system and find the range of parameter values for which it is stable. Consider,

\begin{equation*}
\frac{du}{dt} = a - (1+b)u + u^2 v, \\
\frac{dv}{dt} = bu - u^2v.
\end{equation*}

To find the fixed point, we examine

\begin{equation*}
0 = a - (1+b)u + u^2 v, \\
0 = bu - u^2v,
\end{equation*}

The second equation gives

\begin{equation*}
v = \frac{b}{u}.
\end{equation*}

Substituting this for $v$ into the first equation, we find, 

\begin{equation*}
u^* = a,
\end{equation*}

and therefore

\begin{equation*}
v^* = \frac{b}{a}.
\end{equation*}

The Jacobian is given by

\begin{equation*}
J(u,v) = 
\begin{bmatrix}
-(1+b)+2uv & u^2 \\
b-2uv & -u^2
\end{bmatrix}
\end{equation*}

Evaluating at the fixed point, we have 

\begin{equation*}
J(u^*,v^*) = 
\begin{bmatrix}
b-1 & a^2 \\
-b & -a^2
\end{bmatrix}
\end{equation*}

Thus,

\begin{equation*}
\text{trace}(J) = b-1-a^2\\
\text{det}(J) = a^2(1-b)+a^2b = a^2
\end{equation*}

For stability, we require that $\text{trace}(J)<0$ and $\text{det}(J)>0$. Since $\text{det}(J)=a^2$, the condition $\text{det}(J)>0$ is satisfied for all parameter values. 

The condition $\text{trace}(J)<0$ gives $b<a^2+1.$

\item We now apply the conditions to ensure that $Re(\lambda)>0$. For $D_1J_{22}+D_{2}J_{11}>0$, we first notice that $J_{22} = -a^2$ and $J_{11}= b-1$. This gives $d>a^2/(b-1),$ or $b > 1 + a^2/d$.

Additionally, we need

\begin{equation*}
(D_2J_{11} + D_{1}J_{22})^2 - 4D_1D_2\text{det}(J)>0.
\end{equation*}

Thus

\begin{equation*}
\left(d(b-1) - a^2\right)^2 - 4da^2 > 0 ,
\end{equation*}

which gives 

\begin{equation*}
b > \left(1+\frac{a}{\sqrt{d}}\right)^2
\end{equation*}

Therefore, the range of $b$ for which we have Turing instabilities is 
\begin{equation*}
\left(1+ \frac{a}{\sqrt{b}}\right)^2 < b < 1+a^2
\end{equation*}

\item The range of $k^2$ for which we have instability is 

\begin{equation*}
k_{\pm}^2  = \frac{(d(1+b) -a^2) \pm \sqrt{(d(1+b) -a^2)^2 - 4da^2}}{2d}
\end{equation*}

At the bifurcation, we have $k_c^2 = a/\sqrt{d}$.

\end{enumerate}

## 4. Stripes vs. Spots

For instability in a rectangular domain, 

\begin{equation*}
k_{-}^2 < \left(\frac{n\pi}{L}\right)^2 + \left(\frac{m\pi}{H}\right)^2 < k_{+}^2.
\end{equation*}

Dividing by $(\pi/L)^2$, we have that 

\begin{equation*}
\left( \frac{k_{-}L}{\pi} \right)^2 < n^2 + m^2\left( \frac{L}{H} \right)^2 < \left( \frac{k_{+}L}{\pi} \right)^2 
\end{equation*}

To have stripes, only the mode $n^2 + \left( \frac{L}{H} \right)^2$ must be stable. Therefore, we have 

\begin{equation*}
\frac{L}{H} > \sqrt{\left(\frac{k_{+}L}{\pi}\right)^2 - n^2}.
\end{equation*}


## 5. Zebra stripes

\begin{equation*}
\nabla^2 \phi + k^2 \phi = 0
\end{equation*}

In spherical coordinatates, for the surface with $\theta = \alpha$, we have 

\begin{equation*}
\frac{1}{r^2 \sin ^2 \alpha} \frac{\partial^2 \phi}{\partial \eta ^2} + \frac{1}{r^2}\frac{\partial}{\partial r} \left( r^2 \frac{\partial \phi}{\partial r} \right) + k^2 \phi = 0
\end{equation*}
with conditions 
\begin{equation*}
\frac{\partial \phi}{\partial r}(L,\eta) = 0 \\
\text{ and } \phi < \infty \text{ at } r = 0.  
\end{equation*}

\begin{enumerate}

\item Taking $\phi = \psi(r)/r$, we see
\begin{align*}
\frac{\partial^2 \phi}{\partial \eta^2} &= 0 \\
\frac{\partial \phi}{\partial r} &= \frac{\psi '}{r} - \frac{\psi}{r^2} \\
r^2 \frac{\partial \phi}{\partial r} &= r\psi ' - \psi \\
\frac{\partial}{\partial r} \left( r^2 \frac{\partial \phi}{\partial r} \right) &= \psi' + r \psi'' - \psi' \\
\frac{1}{r^2} \frac{\partial}{\partial r} \left( r^2 \frac{\partial \phi}{\partial r} \right) &= \frac{\psi ''}{r}
\end{align*}

Thus, the Helmholtz equation becomes,

\begin{equation*}
\frac{d^2 \psi}{dr^2} + k^2 \psi = 0,
\end{equation*}

the general solution of which is

\begin{equation*}
\psi(r) = A\cos kr + B\sin kr.
\end{equation*}


\item For $\phi<\infty$ at $r = 0$, we must have $A=0$.

For the other condition, we first compute

\begin{equation*}
\frac{\partial}{\partial r}\left[ \frac{B\sin kr}{r} \right] = \frac{Bk\cos kr}{r} - \frac{B\sin kr}{r^2}
\end{equation*}

Evaluating this at $r=L$ gives

\begin{equation*}
kL\cos kL - \sin kL = 0, \\
\text{ or} \\
kL = \tan kL.
\end{equation*}

The smallest value for which this equation is satisfied is 

\begin{equation*}
k \approx 4.5/L.
\end{equation*}


\end{enumerate}

## 6. Patterns in a petri dish


\begin{enumerate}

\item Taking $\phi(r,\theta) = \psi(r)f(\theta)$ into $\nabla^2\phi + k^2\phi = 0$ gives

\begin{equation*}
\frac{f}{r} \frac{d}{dr}\left( r \frac{d\psi}{dr} \right) + \frac{\psi}{r^2} \frac{d^2 f}{d\theta^2} + k^2 f\psi = 0
\end{equation*}

Dividing by $\frac{f\psi}{r^2}$, we have

\begin{equation*}
\frac{r}{\psi}\frac{d}{dr}\left ( r\frac{d\psi}{dr} \right) + k^2r^2 = -\frac{1}{f} \frac{d^2f}{d\theta^2}
\end{equation*}

Since the LHS is a function of $r$ and the RHS is a function of $\theta$, the only way these can be equal is if they are constant. Calling this constant $m^2$, we have

\begin{equation*}
\frac{d^2f}{d\theta^2} + m^2 f =0 \\
r^2 \frac{d^2\psi}{dr^2} + r\frac{d\psi}{dr} + (r^2 k^2 -m^2) =0
\end{equation*}

\item The solution to the equation for $f$ is 
\begin{equation*}
f(\theta) = A\cos(m\theta) + B\sin(m\theta).
\end{equation*}
The condition that $f$ is $2\pi$-periodic requires that $m$ be an integer. 

\item Rescaling such that $\rho = kr$ gives Bessel's equation

\begin{equation*}
\rho^2 \frac{d^2 \psi}{d\rho^2} + \rho \frac{d\psi}{d\rho} + (\rho^2 - m^2)\psi = 0
\end{equation*}

whose solutions are Bessel functions

\begin{equation*}
\psi = J_m(\rho)
\end{equation*}

for each integer $m$.

\item From the condition that

\begin{equation*}
\left.\frac{d\phi}{dr}\right|_{r = R} = 0,
\end{equation*}

we have,

\begin{equation*}
\left.\frac{d\psi}{dr}\right|_{r = R} = \left.k \frac{dJ_m}{d\rho}\right|_{\rho = kR} = 0.
\end{equation*}

For the case of concentric rings, $\phi$ will not depend on $\theta$. Hence, $f = $ const. and therefore $m^2 = 0$ and $\psi(r) = J_0(kr)$.  The boundary condition requires that

\begin{equation*}
\left.\frac{dJ_0}{d\rho} \right|_{\rho = kR} = -J_1(kR) = 0
\end{equation*}

We see then that the values of $k$ are determined by the values of $kR$ where $J_1(kR) = 0$. For $J_0(\rho)$ to have two minima, we take the zero of $J_1$ to be $kR = 13.3237$. A value of $R = 4$ ensures $3 < k < 3.5$ (other choices are possible). 

\end{enumerate}

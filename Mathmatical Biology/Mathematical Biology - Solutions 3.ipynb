## 1. Glycolysis

\begin{enumerate}


\item For each reaction, we have

\begin{itemize}

\item \begin{equation*}
\emptyset \xrightarrow{\beta}Y \\
\frac{dy}{dt} = \beta
\end{equation*}


\item \begin{equation*}
Y \xrightarrow{\alpha}X \\
\frac{dy}{dt} = -\alpha y \\
\frac{dx}{dt} = \alpha y
\end{equation*}

\item \begin{equation*}
2X + Y \xrightarrow{\gamma}3X \\
\frac{dy}{dt} = -\gamma yx^2 \\
\frac{dx}{dt} = \gamma yx^2
\end{equation*}

\item \begin{equation*}
X \xrightarrow{\mu} \emptyset \\
\frac{dx}{dt} = -\mu x
\end{equation*}

\end{itemize}

Adding the resulting equations together, we have that

\begin{equation*}
\frac{dx}{dt} = \alpha y + \gamma y x^2 - \mu x \\
\frac{dy}{dt} = \beta - \alpha y -\gamma y x^2
\end{equation*}


\item To nondimensionalise, we take 

\begin{equation*}
t = \frac{\tau}{\mu}\, , \, x = \left(\frac{\mu}{\gamma}\right)^{1/2} u \, , \, y = \left(\frac{\mu}{\gamma}\right)^{1/2} v
\end{equation*}

Substituting the expressions into the differential equations gives:

\begin{equation*}
\frac{du}{d\tau} = av + vu^2 -u \\
\frac{dv}{d\tau} = b - av - vu^2
\end{equation*}

where \begin{equation*}
a = \frac{\alpha}{\mu}\, , \, \, b = \left(\frac{\gamma}{\mu}\right)^{1/2} \beta
\end{equation*}


\item The nullclines are given by

\begin{equation*}
f_1(u,v) = -u + av + vu^2 = 0 \\
f_2(u,v) = b - av - vu^2 = 0
\end{equation*}

Thus, from $f_1 = 0$, we have 
\begin{equation*}
v = \frac{u}{a+u^2}
\end{equation*}
and from $f_2 = 0$,
\begin{equation*}
v = \frac{b}{a+u^2}
\end{equation*}

The nullclines are plotted shown below and are generated using the Python cell.

\item The fixed points are given by 

\begin{equation*}
\frac{b}{a+u^2} = \frac{u}{a+u^2},
\end{equation*}
and therefore,
\begin{equation*}
(b-u)(a+u^2) = 0.
\end{equation*}
Since $a>0$, we only have one fixed point at 
\begin{equation*}
(u^*, v^*) = \left(b, \frac{b}{b^2+a} \right)
\end{equation*}


\item By considering each boundary in turn, we can confirm that the trapping set conditions are satisfied.

\begin{itemize}

\item For the vertical line connecting $(0, 0)$ and $(0, b/a)$, we have $u = 0$, and therefore $\hat{\mathbf{n}} = - \hat{\mathbf{u}}$.

\begin{equation*}
\mathbf{f}(0,v)\cdot\hat{\mathbf{n}} = -f_1(0,v) = -av
\end{equation*}

Thus, $\mathbf{f}\cdot\mathbf{\hat{n}} \leq 0$ for $v\geq0$. 

\item For the jorizontal line connecting $(0, b/a)$ and $(b, b/a)$, $v = b/a$, and $\hat{\mathbf{n}} = \hat{\mathbf{v}}$.

\begin{equation*}
\mathbf{f}(u,b/a)\cdot\mathbf{\hat{n}} = f_2(u,b/a) \\
f_2(u,b/a) = b-a(b/a)-u^2(b/a) = -u^2\frac{b}{a} \leq 0
\end{equation*}

\item For the diagonal line of slope âˆ’1 from the point $(b,b/a)$ to the point where it intersects the $u$-nullcline:

Since the boundary is a line with slope -1, we have that $\sqrt{2}\mathbf{\hat{n}} = \mathbf{\hat{u}} + \mathbf{\hat{v}}$.

Thus, 
\begin{equation*}
\sqrt{2}\mathbf{f}\cdot\mathbf{\hat{n}} = f_1 + f_2 = -u + b.
\end{equation*}
Since $u\geq b$, we then have $\mathbf{f}\cdot\mathbf{\hat{n}} \leq 0$.

\item For the vertical line from $(u_0, v_0)$ to $(u_0, 0)$, $\mathbf{\hat{n}} = \mathbf{\hat{u}}$.  

\begin{equation*}
\mathbf{\hat{n}}\cdot\mathbf{f} = f_1
\end{equation*}

This line sits below the nullcline where $f_1\leq0$. Thus, $\mathbf{\hat{n}}\cdot\mathbf{f}\leq 0$.

\item For horizontal line from $(u_0, 0)$ to $(0, 0)$, $v = 0$ and $\mathbf{\hat{n}} = - \mathbf{\hat{v}}$.

\begin{equation*}
\mathbf{f}\cdot\mathbf{\hat{n}}  = -f_2(u,v) = -b < 0.
\end{equation*}

\end{itemize}

\item To puncture the set, we need to make sure that the fixed point is an unstable node or unstable spiral. The Jacobian is given by 

\begin{equation*}
J(u,v) = 
\begin{bmatrix}
-1 + 2uv & a+u^2 \\
-2uv & -(a+u^2)
\end{bmatrix}
\end{equation*}

For $u = u^*$, $v = v^*$ we have that

\begin{equation*}
J(u^*,v^*) = 
\begin{bmatrix}
-1 + \frac{2b^2}{b^2+a} & a+b^2 \\
\frac{-2b^2}{b^2 + a} & -(a+b^2)
\end{bmatrix}.
\end{equation*}

The determinant of $J$ is given by 

\begin{equation*}
\text{det}(J) = a+b^2-2b^2+2b^2 = a+b^2 > 0
\end{equation*}

while the trace is 

\begin{equation*}
\text{trace}(J) = - 1 - (a+b^2) + \frac{2b^2}{b^2+a} = b^2 - a -(b^2+a)^2.
\end{equation*}

In order to have an unstable node or spiral, we require $trace(J)>0$. Thus, $b^2$ must be lie between the roots of 

\begin{equation*}
b^4 + b^2(2a-1) + a(a+1) = 0
\end{equation*}

The roots are

\begin{equation*}
b_{\pm} = \frac{1}{2} (1-2a\pm\sqrt{1-8a})
\end{equation*}

and therefore we must have $b_{-}< b <b_{+}$ and $a\leq\frac{1}{8}$

\end{enumerate}

## 2. Period of limit cycles in predator-prey systems

__Limit cycle I:__

We first find the fixed point by considering

\begin{equation*}
0 = u(1-u)-\frac{auv}{u+d} \\
0 = bv\left(1-\frac{v}{u}\right)
\end{equation*}

From the second equation we have that $u = v$. Substituting this into the first equation, we find that 

\begin{equation*}
u(u-1) = \frac{au^2}{u+d}.
\end{equation*}

Thus, 

\begin{equation*}
u[(1-u)(u+d)-au] = 0,
\end{equation*}

and we consider,

\begin{equation*}
u^2 + (a+d-1)u-d = 0,
\end{equation*}

which yields the value of interest

\begin{equation*}
u^* = \frac{(1-a-d)+\sqrt{(a+d-1)^2+4d}}{2}
\end{equation*}

The Jacobian evaluated at the fixed point (see notes) is 

\begin{equation*}
J(u^*,v^*) = 
\begin{bmatrix}
u^*(\frac{au^*}{(u^*+d)^2}-1) & \frac{-au^*}{u^* + d} \\
b & -b
\end{bmatrix}
\end{equation*}

Let
\begin{equation*}
\alpha = u^* (\frac{au^*}{(u^*+d)^2}-1)\,, \\
\beta = \frac{-au^*}{u^* + d}.
\end{equation*}

Thus, $det(J - \lambda I) = 0$ gives

\begin{equation*}
-(\alpha-\lambda)(b+\lambda) + b \beta = 0,
\end{equation*}

and further, 

\begin{equation*}
\lambda^2 + (b-\alpha)\lambda - b (\alpha-\beta) = 0
\end{equation*}

which has roots

\begin{equation*}
\lambda = \frac{-(b-\alpha)\pm\sqrt{(b-\alpha)^2+4b(\alpha-\beta}}{2}.
\end{equation*}

Thus, the Hopf bifurcation occurs at $b = \alpha$ with
\begin{equation*}
\lambda = \pm \sqrt{b(\alpha-\beta)}.
\end{equation*}

We see that 

\begin{equation*}
\alpha - \beta = u^*\left( \frac{\beta}{u^* + d} -1 \right) - \beta \\
 = \frac{-\beta d}{u^*+d}-u^* <0
\end{equation*}

since $\beta > 0$, $u^*>0$, $d>0$. Thus we can express the eigenvalue at the bifurcation as

\begin{equation*}
\lambda = \pm i \omega \text{ where } \omega = \sqrt{b(\beta - \alpha)}
\end{equation*}

and

\begin{equation*}
T = \frac{2\pi}{\sqrt{b(\beta - \alpha)}}
\end{equation*}


__Limit cycle II__

\begin{enumerate}

\item  

Taking $b = \epsilon << 1$, we know then that $\frac{dv}{d\tau} \approx 0$ while $u$ relaxes to its steady state value given by $f_1(u,v) = 0$.

This will occur provided $f_1$ and $f_2$ remain $O(1)$. 

\item As $f_2 > 0$ in this region of the phase space, the trajectory will move up the nullcline until reaching the local maximum. 

In the 1D system described by 

\begin{equation*}
\frac{du}{d\tau} = f_1(u;v),
\end{equation*}

this corresponds to the point where a saddle-node bifurcation occurs, causing the steady solution to jump to another stable branch.

Thinking back again about the system, we know that $f_2 > 0$ in this region and, therefore, the solution cannot follow the nullcline downwards. 

\item When $u$ becomes small, we see that $f_2$ no longer remains $O(1)$. 

As the trajectory is nearly horizontal with value $v = v_2$, $f_2$ becomes $O(1/\epsilon)$ when 

\begin{equation*}
\frac{v_2^2}{u} \approx \frac{1}{\epsilon} \text{ or } u \approx \epsilon v_2^2 
\end{equation*}
 

\item For this portion of the orbit, we scale $u$ such that $u = \epsilon U$ where $U$ is $O(1)$. Substituting this into our equations, we obtain

\begin{equation*}
\frac{dU}{d\tau} = U(1-\epsilon U)-\frac{avU}{d+\epsilon U}\\
\frac{dv}{d\tau} = \epsilon v ( 1 - \frac{v}{\epsilon U})
\end{equation*}

For $\epsilon \ll 1$, we have to leading order that

\begin{equation*}
\frac{dU}{d\tau} = U - \frac{avU}{d} \\
\frac{dv}{d\tau} = -\frac{v^2}{U}.
\end{equation*}

We see then that there is no longer a separation of timescales and both $U$ and $v$ change at fast time, $\tau$. 

\item As a result, the period can be approximated by the time spent on the nullcline. Thus, we consider 

\begin{equation*}
\int_{v_1}^{v_2} \frac{dv}{v(1-v/u)} = \epsilon T \\
\int_{v_1}^{v_2} \frac{u}{v(u-v)}dv = \epsilon T \\
\int_{u_1}^{u_2} \frac{u}{v(u-v)}\left(\frac{dv}{du}\right)du = \epsilon T \\
\end{equation*}

Since on the nullcline we have 

\begin{equation*}
v(u) = \frac{(1-u)(u+d)}{a}, \\
a \frac{dv}{du} = 1 - d 2u,
\end{equation*}

and  the period is given by 

\begin{equation*}
T = \frac{1}{a \epsilon} \int_{u_1}^{u_2} \frac{u(1-d-2u)}{v(u)[u-v(u)]} du
\end{equation*}

\end{enumerate}

## 3. Genetic toggle switch

\begin{enumerate}

\item 


The rate equations are 

\begin{align*}
	\frac{d G_1}{d t} &= - \beta_1 G_1 D_2 + \beta_2 G_1^\ast, \\ 
    \frac{d G_2}{d t} &= - \beta_1 G_2 D_1 + \beta_2 G_2^\ast, \\
	\frac{d G_1^\ast}{d t} &= - \frac{d G_1}{d t}, \\
    \frac{d G_2^\ast}{d t} &= - \frac{d G_2}{d t},  \\
	\frac{d D_1}{d t} &=  \alpha_1 P_1^2 - \alpha_2 D_1 + \frac{d G_2}{d t}, \\ 
    \frac{d D_2}{d t} &=   \alpha_1 P_2^2 - \alpha_2 D_2 + \frac{d G_1}{d t} \\
	\frac{d P_1}{d t} &= - 2 \alpha_1 P_1^2 + 2\alpha_2 D_1 + k_0 G_1 - k_1 P_1, \\
	\frac{d P_2}{d t} &= - 2 \alpha_1 P_2^2 + 2\alpha_2 D_2 + k_0 G_2 - k_1 P_2
\end{align*}

\item First, we note that 

\begin{align*}
G_1+G_1^\ast = const = g_1, \ \ G_2+G_2^\ast = const = g_2.
\end{align*}

Setting the time-derivative of the first five equations to zero, we obtain

\begin{align*}
	G_1 &= g_1 \frac{K^2}{K^2+{P_2}^2}, \\ 
    G_2 &= g_2 \frac{K^2}{K^2+{ P_1}^2}
\end{align*}

where $K^2=\frac{\beta_2}{\beta_1} \frac{\alpha_2}{\alpha_1}.$ Then,

\begin{align*}
	\frac{d P_1}{d t} &= k_0 g_1 \frac{K^2}{P_2^2+K^2} - k_1 P_1, \\

\end{align*}

Setting $\tau=k_1 t$, $x=P_1/K$ and $y=P_2/K$, we find the non-dimensional equations

\begin{align}
 \frac{d x}{d \tau} &= \frac{a_1}{1+y^2} - x, \\
 \frac{d y}{d \tau} &= \frac{a_2}{1+x^2} - y. 
 \label{eq:toggle}

where $a_{n}= k_0g_n/(k_1K).$

## 4. The Brusselator

\begin{enumerate}

\item For each reaction we have:

\begin{itemize}

\item
\begin{equation*}
A \xrightarrow{k_1} X \\
\frac{dx}{dt} = k_1 a
\end{equation*}

\item \begin{equation*}
B+X \xrightarrow{k_2} Y+D \\
\frac{dx}{dt} = -k_2 xb \\
\frac{dy}{dt} = k_2 x b
\end{equation*}

\item \begin{equation*}
2X + Y \xrightarrow{k_3} 3X \\
\frac{dx}{dt} = k_3 y x^2 \\
\frac{dy}{dt} = -k_3 y x^2
\end{equation*}

\item \begin{equation*}
X \xrightarrow{k_4} E \\
\frac{dx}{dt} = -k_4 x
\end{equation*}

\end{itemize}

Summing these equations together gives,

\begin{align*}
\frac{dx}{dt} &= k_1 a - k_2 bx + k_3 y x^2 - k_4 x, \\
\frac{dy}{dt} &= k_2 b x - k_3 y x ^2.
\end{align*}


\item 
To nondimensionalise, we introduce

\begin{equation*}
t = \tau/k_4, \, x = \frac{k_1 a}{k_4}u, \, y = \frac{k_1 a}{k_4} v.
\end{equation*}

This gives, 

\begin{equation*}
\frac{du}{d\tau} = 1 - (\beta + 1)u + \alpha u^2 v, \\
\frac{dv}{d\tau} = \beta u - \alpha u^2 v,
\end{equation*}

where,

\begin{equation*}
\beta = \frac{k_2 b}{k_4},\, \alpha = \frac{k_3 (k_1 a )^2}{k_4^3}.
\end{equation*}

\item We can find the fixed point by considering,

\begin{equation*}
0 = 1 - (1 + \beta)u + \alpha u^2v,\\
0 = \beta u -\alpha u^2v.
\end{equation*}

The second equation above yields,

\begin{equation*}
u = 0 \text{ and } v = \frac{\beta}{\alpha u}.
\end{equation*}

Substituting the second expression into the first equation, we find,  

\begin{equation*}
\frac{\beta}{\alpha u} = \frac{(1+\beta)u - 1}{\alpha u^2},
\end{equation*}

and solving for $u$, we see 

\begin{equation*}
0 = u(u-1).
\end{equation*}

Thus,

\begin{equation*}
u^* = 1 \text{ and } v^* = \beta/\alpha 
\end{equation*}

To find the Jacobian, we first compute the partial derivatives,

\begin{align*}
\frac{\partial f_1}{\partial u} &= -(\beta + 1) + 2\alpha uv \\
\frac{\partial f_1}{\partial v} &= \alpha u^2 \\
\frac{\partial f_2}{\partial u} &= \beta - 2\alpha uv \\
\frac{\partial f_2}{\partial v} &= -\alpha u^2 
\end{align*}

Thus,

\begin{equation*}
J(1, \beta/\alpha) = 
\begin{bmatrix}
\beta - 1 & \alpha \\
-\beta & -\alpha
\end{bmatrix}
\end{equation*}

To compute the eigenvalues, we consider,

\begin{equation*}
det(J-\lambda I) = \lambda ^2 + \lambda(\alpha + 1 - \beta) + \alpha = 0,
\end{equation*}

to find,

\begin{equation*}
\lambda  = \frac{-(\alpha + 1 - \beta)\pm \sqrt{(\alpha + 1 - \beta)^2 - 4\alpha}}{2}.
\end{equation*}

From this expression, we see that the bifurcation will occur when 

\begin{equation*}
\alpha + 1 - \beta = 0
\end{equation*}

And thus, $\beta = 1 + \alpha$.  We also want to ensure that the fixed point goes from a stable to unstable spiral at the bifurcation.  We then consider,

\begin{align*}
(\alpha + 1 - \beta)^2 - 4 \alpha &< 0 \\
(\alpha + 1 - \beta)^2 &< 4\alpha \\
-2\sqrt{\alpha} &< \alpha + 1 -\beta < 2 \sqrt{\alpha}
\end{align*}

and we will have a spiral when we pass through the bifurcation.

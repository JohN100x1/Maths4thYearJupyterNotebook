# 1. Mean and variance of the Poisson process

The mean and variance are given by 

\begin{align*}
m(t) &= \frac{\partial \mathcal{P}}{\partial z}\Big{|}_{z = 1}, \\
\sigma^2(t) &= \frac{\partial^2 \mathcal{P}}{\partial z^2}\Big{|}_{z = 1} +  \frac{\partial \mathcal{P}}{\partial z}\Big{|}_{z = 1} - \left(\frac{\partial \mathcal{P}}{\partial z}\Big{|}_{z = 1}\right)^2.
\end{align*}

For 

\begin{align*}
\mathcal{P}(z,t) = e^{\lambda (z - 1) t}
\end{align*}

we have

\begin{align*}
\frac{\partial \mathcal{P}}{\partial z} = \lambda t e^{\lambda(z-1)t} \\
\frac{\partial^2 \mathcal{P}}{\partial z^2} = (\lambda t)^2 e^{\lambda(z-1)t}
\end{align*}

Thus

\begin{align*}
m(t) &= \frac{\partial \mathcal{P}}{\partial z}\Big{|}_{z = 1} = \lambda t, \\
\sigma^2(t) &= (\lambda t)^2 + \lambda t - (\lambda t)^2 \\
&= \lambda t
\end{align*}

# 2. Variance of the simple birth process

In the notes, we showed that

\begin{align*}
m = Ne^{\lambda t}.
\end{align*}

To determine the variance, we multiply the forward Kolmogorov equation by $i^2$ and sum to find

\begin{equation}\label{eq:q2}
\sum_{i=N}^\infty i^2 \frac{dp_i}{dt} = \sum_{i=N+1}^{\infty} \lambda i^2 (i-1) p_{i-1} - \sum_{i=N}^{\infty} \lambda i^3 p_i
 \end{equation}

We see that 

\begin{equation*}
\sum_{i=N}^{\infty} i^2 \frac{dp_i}{dt} = \frac{d}{dt} \sum_{i=N}^\infty i^2 p_i \\
\sum_{i = N+1}^{\infty}  \lambda i^2 (i-1) p_{i-1} = \sum_{i=N}^\infty \lambda (i+1)^2 ip_i
\end{equation*}

The RHS of \eqref{eq:q2} is then

\begin{equation*}
\sum_{i=N}^\infty \lambda (i+1)^2 ip_i - \sum_{i=N}^{\infty} \lambda i^3 p_i \\
= 2 \lambda \sum_{i=N}^\infty i^2 p_i + \lambda \sum_{i=N}^\infty ip_i
\end{equation*}

Defining $\gamma = \sum_{i=N}^\infty i^2 p_i$, we have

\begin{equation*}
\frac{d\gamma}{dt} = 2\lambda \gamma + \lambda m \\
\frac{d\gamma}{dt} - 2\lambda \gamma = \lambda N e^{\lambda t}
\end{equation*}

The general solution is

\begin{equation*}
\gamma = \gamma_0 e^{2\lambda t} - Ne^{\lambda t}
\end{equation*}

Since 

\begin{align*}
\sigma^2 &= \sum i^2 p_i - m^2 \\
&= \gamma -m^2 \\
&= \gamma_0 e^{2\lambda t} - Ne^{\lambda t} - N^2e^{2\lambda t} \\
&= (\gamma -N^2)e^{2\lambda t} - Ne^{\lambda t}
\end{align*}

At $t = 0$ we know $X(0) = N$ and thus $\sigma^2(0) = 0$.

As a result, 

\begin{equation*}
0 = (\gamma_0 -N^2)-N \\
\gamma_0 = N(N+1) \\
\sigma^2(t) = N[e^{2\lambda t} - e^{\lambda t}] \\
= Ne^{\lambda t}[e^{\lambda t}-1]
\end{equation*}



# 3. Simple death process
\begin{enumerate}

\item From the transition probabilities, we see that the non-zero entries of the generator matrix are

\begin{align*}
q_{i,i} &= -\mu i \\
q_{i-1, i} &= \mu i
\end{align*}

The forward Kolmogorov equation is then 
\begin{equation*}
\frac{dp_i}{dt} = -\mu i p_i + \mu(i+1)p_{i+1}
\end{equation*}

Multiplying by $z^i$ and summing, we have

\begin{equation*}
\frac{d}{dt} \sum_{i} p_i z^i = -\mu\sum_{i} ip_iz^i + \mu \sum_{i} (i+1) p_{i+1} z^i
\end{equation*}

We can rewrite the sums on the RHS as

\begin{align*}
\sum_{i} i p_i z^i &= z \sum_{i} i p_i z^{i-1} \\
  &= z\frac{\partial }{\partial z} (\sum_{i} p_i z^i )
\end{align*}

and

\begin{align*}
\sum_{i}(i+1)p_{i+1}z^i &= \sum_{i} i p_i z^{i-1} \\
&= \frac{\partial}{\partial z} \sum_{i} p_i z^i
\end{align*}

Since $\mathcal{P}(z,t) = \sum_{i} p_i z^i$ we have that

\begin{equation*}
\frac{\partial \mathcal{P}}{\partial t} = \mu (1-z) \frac{\partial \mathcal{P}}{\partial z}
\end{equation*}

\item Introducing $\tau$ in the usual way, we have

\begin{align*}
\frac{d \mathcal{P}}{d \tau} &= 0, \\
\frac{dt}{d\tau} &= 1, \\
\frac{dz}{d\tau} &= \mu (z-1),
\end{align*}

The first two equations give 

\begin{equation*}
\mathcal{P}=c_1\,, t = \tau + c_2
\end{equation*}

The third equation is separable and we see that 

\begin{align*}
\int \frac{dz}{z-1} &= \int \mu \tau \\
\log(z-1) &= \mu \tau + c \\
z &= c_3 e^{\mu \tau} + 1.
\end{align*}

Since the initial data at $\tau=0$ is given on the $z$ axis which should be parameterised by $s$, we have

\begin{align*}
\mathcal{P}(s,0) &= s^N, \\
t &= \tau, \\
s &= c_3+1
\end{align*}

The third equation gives $c_3 = s-1$.

Using this in the expression for $z$, we have

\begin{equation*}
z = (s-1) e^{\mu\tau} +1 .
\end{equation*}

Solving for $s$, we see that 

\begin{equation*}
s = 1+(z-1)e^{-\mu t}
\end{equation*}

Thus, 

\begin{equation*}
\mathcal{P}(z,t) = [1+(z-1)e^{-\mu t}]^N
\end{equation*}

\item From $\mathcal{P}(z,t)$, we can compute $m$ and $\sigma^2$.

We have 

\begin{align*}
\frac{\partial \mathcal{P}}{\partial z} &= N[1+e^{-\mu t}(z-1)]^{N-1}e^{-\mu t} \\
\frac{\partial ^2 \mathcal{P}}{\partial z^2} &= N(N-1)[1+ e^{-\mu t}(z-1)]^{N-2} e^{-2\mu t}
\end{align*}

and therefore

\begin{equation*}
m(t) = Ne^{-\mu t}
\end{equation*}

and 

\begin{align*}
\sigma^2(t) &= N(N-1)e^{-2\mu t}+Ne^{-\mu t}-N^2e^{-2\mu t} \\
&= Ne^{-\mu t}(1-e^{-\mu t}).
\end{align*}



\end{enumerate}

# 4. Simple birth and death process with immigration

From the transition probabilities, we see that the entries of the generator matrix are 

\begin{equation*}
q_{i,i} = -\nu + (\lambda + \mu)i \\
q_{i+1, i}= \gamma + \lambda i \\
q_{i-1,i} = \mu i
\end{equation*}

As a result, the forward Kolmogorov equation is

\begin{equation*}
\frac{dp_i}{dt} = -[\nu + (\lambda + \mu)i]p_i + [\nu + \lambda(i-1)]p_{i-1} + \mu (i+1)p_{i-1}
\end{equation*}

We can find the equation for the mean by multiplying the above equation by $i$ and summing such that

\begin{equation*}
\sum_i i\frac{dp_i}{dt} = -\sum_i i(\nu + (\mu + \lambda)i)p_i + \sum_i i (\nu + \lambda(i-1))p_{i-1} + \sum_i \mu (i+1)p_{i+1}
\end{equation*}

We can rewrite two of the sums on the RHS as 

\begin{equation*}
\sum_i i (\nu + \lambda (i-1))p_{i-1} \\
= \sum_i [\nu + \lambda i] (i+1) p_i
 \end{equation*}

and 

\begin{equation*}
\sum_i i\mu(i+1)p_{i+1} = \sum_i i(i-1)\mu p_i.
\end{equation*}

Using these results and the fact that 

\begin{equation*}
\sum_i i \frac{dp_i}{dt} = \frac{dm}{dt}
\end{equation*}

we have

\begin{equation*}
\frac{dm}{dt} = \nu + \lambda m - \mu m.
\end{equation*}

This has the general solution 

\begin{equation*}
m = ke^{(\lambda-\mu)t} + \frac{\nu}{\mu-\lambda}
\end{equation*}

Using the initial condition, $m(0) = N$, we find

\begin{equation*}
k = N+\frac{\nu}{\lambda + \mu}
\end{equation*}

and therefore

\begin{equation*}
m(t) = Ne^{(\lambda - \mu)t} + \frac{\nu}{\lambda - \mu}[e^{(\lambda - \mu)t}-1]
\end{equation*}

Based on the rates $\nu, \lambda N, -\mu N$, we have the equation 

\begin{equation*}
\frac{dN}{dt} =\nu + (\lambda - \mu) N
\end{equation*}

which is precisely the equation for the mean. 


# 5. Burst process

\begin{enumerate}

\item

Since the next generation generating function is 
\begin{equation*}
f(z) = z^{\beta + 1},
\end{equation*}
the backward Kolmogorov differential equation is 
\begin{equation*}
\frac{d\mathcal{P}}{dt}= -\lambda [\mathcal{P} - \mathcal{P}^{\beta+1}]\\
= \lambda\mathcal{P}[\mathcal{P}^{\beta}-1]
\end{equation*}

\item This equation is seperable and accordingly, 
\begin{equation*}
\int \frac{d\mathcal{P}}{\mathcal{P}[\mathcal{P}^{\beta}-1]} = \int \lambda dt \\
\frac{1}{\beta} \log(1-\mathcal{P}^{\beta})-\log \mathcal{P} = \lambda t + c_0 \\
\frac{1-\mathcal{P}^{\beta}}{\mathcal{P}^{\beta}} = ke^{\beta \lambda t}
\end{equation*}
Since at $t=0$, $\mathcal{P} = z$, 
\begin{equation*}
\frac{1-z^{\beta}}{z^{\beta}} = k
\end{equation*}
And therefore
\begin{equation*}
\frac{1-\mathcal{P}^{\beta}}{\mathcal{P}^{\beta}} = \frac{1-z^\beta}{z^\beta}e^{\beta \lambda t}.
\end{equation*}
Solving for $\mathcal{P}$, we find 
\begin{equation*}
\mathcal{P}(z,t) = [1+e^{\beta \lambda t}(z^{-\beta}-1)]^{\frac{-1}{\beta}}
\end{equation*}

\item

The mean is given by 
\begin{align*}
m(t) &= \frac{\partial \mathcal{P}}{\partial z}\Big{|}_{z = 1}
\end{align*}

As
\begin{equation*}
\frac{\partial \mathcal{P}}{\partial z} = \frac{e^{\beta \lambda t}z^{-(\beta -1)}}{[1+ e^{\beta \lambda t}(z^{-\beta} -1)]^{1+\frac{1}{\beta}}}
\end{equation*}
we have
\begin{equation*}
m(t) = e^{\beta \lambda t}
\end{equation*}


\end{enumerate}

# 6. Bivariate birth and death process

The forward Kolmogorov equation for this process is given in the notes and is 

\begin{align*}
\frac{d p_{(m,n)}}{d t} &= - p_{(m,n)}(t) \left((\lambda_1 + \mu_1)m + (\lambda_2 + \mu_2)n\right) \\
&+ \lambda_1(m-1)p_{(m-1,n)}(t) + \lambda_2(n-1)p_{(m,n-1)}(t)\\
&+ \mu_1(m+1)p_{(m+1,n)}(t) + \mu_2(n+1)p_{(m,n+1)}(t).
\end{align*}

Let's first find 
\begin{align*}
\mu_X &= \sum_{m = 0}^{\infty}\sum_{n = 0}^{\infty}m p_{(m,n)}(t)
\end{align*}

To do so, we multiply the FKE by $m$ and sum over $(m,n)$ to find
\begin{align}
\sum_{(m,n)} m \frac{d p_{(m,n)}}{dt} &= \frac{d \mu_X}{dt} \nonumber \\
&= - \sum_{(m,n)}p_{(m,n)}(t) m\left((\lambda_1 + \mu_1)m + (\lambda_2 + \mu_2)n\right) \nonumber \\
&+ \sum_{(m,n)}\lambda_1 m (m-1)p_{(m-1,n)}(t) + \sum_{(m,n)}\lambda_2 m(n-1)p_{(m,n-1)}(t) \nonumber \\
&+ \sum_{(m,n)}\mu_1m(m+1)p_{(m+1,n)}(t) + \sum_{(m,n)}\mu_2m(n+1)p_{(m,n+1)}(t) \label{eq:q6}
\end{align}

We can rewrite the following sums as 
\begin{align*}
\sum_{(m,n)}\lambda_1m(m-1)p_{(m-1,n)}(t) &= \sum_{(m,n)}\lambda_1(m+1)mp_{(m,n)}(t) \\
\sum_{(m,n)}\lambda_2m(n-1)p_{(m,n-1)}(t) &= \sum_{(m,n)}\lambda_2 mnp_{(m,n)}(t) \\
\sum_{(m,n)}\mu_1m(m+1)p_{(m+1,n)}(t) &= \sum_{(m,n)}\mu_1(m-1)mp_{(m,n)}(t) \\
\sum_{(m,n)}\mu_2m(n+1)p_{(m,n+1)}(t) &= \sum_{(m,n)}\mu_2mnp_{(m,n)}(t).
\end{align*}

Using these results in \eqref{eq:q6} above, we have
\begin{equation*}
\frac{d \mu_X}{dt} = \lambda_1 \sum_{(m,n)} mp_{(m,n)} - \mu_1\sum_{(m,n)} mp_{(m,n)}
\end{equation*}
and therefore,

\begin{equation*}
\frac{d \mu_X}{dt} = (\lambda_1-\mu_1)\mu_X.
\end{equation*}

By similar arguments, we also have 
\begin{equation*}
\frac{d \mu_Y}{dt} = (\lambda_2-\mu_2)\mu_Y.
\end{equation*}

# 7. Noise propagation in a gene expression
\begin{enumerate}

\item The law of mas action gives the deterministic equations
\begin{equation*}
\frac{dm}{dt} = \underbrace{k_1}_{\lambda_1} - \underbrace{k_m m}_{\mu_1} \\
\frac{dp}{dt} = \underbrace{k_2 m}_{\lambda_2} - \underbrace{k_p p}_{\mu_2}
\end{equation*}

This is a bivariate birth and death process with birth rates $\lambda_1 = \tilde{k}_1$ and $\lambda_2$, and death rates $\mu_1$ and $\mu_2$, as indicated in the equation above.  Accordingly, the transition probabilities correspond to those in question 6 and with the forward Kolmogorov equation given in the notes.  Then, multiplying the forward Kolmogorov differential equation by $m$ and summing, we obtain

\begin{align*}
\frac{d\mu_m}{dt} = &-\sum_{(m,n)}m p_{(m,n)}[\tilde{k}_1 + k_mm + k_2m + k_p n] \\
&+ \sum_{(m,n)}m \tilde{k}_1 p_{(m-1,n)} \\
&+ \sum_{(m,n)}m (m+1)k_m p_{(m+1,n)} \\
&+ \sum_{(m,n)}m^2 k_2 p_{(m,n-1)} \\
&+ \sum_{(m,n)}m(n+1) k_p p_{(m,n+1)} 
\end{align*}

Rearranging in a similar way to the Q6 solutions, we find

\begin{equation*}
\frac{d\mu_m}{dt} = \tilde{k}_1\sum_{(m,n)}p_{(m,n)} - k_m \sum_{(m,n)} mp_{(m,n)} \\
= \tilde{k}_1 -k_m \mu_m
\end{equation*}

We now multiply the forward Kolmogorov equation by $n$, sum over $(m,n)$, and rearrange the sums.  In doing so, we obtain

\begin{equation*}
\frac{d\mu_p}{dt} = k_2\mu_m -k_p\mu_p
\end{equation*}

\item For the variance $\sigma^2_m$, we multiply the FK equation by $m^2$ and sum over $(m,n)$.  We then rearrange and combine the terms to give

\begin{equation*}
\frac{d}{dt} \sum_{(m,n)}m^2 p_{(m,n)} = 2\tilde{k}_1 \mu_m + \tilde{k}_1-k_m \sum_{(m,n)}m^2 p_{(m,n)} \\
\end{equation*}

Thus for $\sigma^2_m$, we have that

\begin{equation*}
\frac{d\sigma^2}{dt} = \frac{d}{dt}(\sum_{(m,n)} m^2 p_{(m,n)}) -2\mu_m \frac{d\mu_m}{dt} \\
\frac{d\sigma^2_m}{dt} = \tilde{k}_1 - 2k_m \sigma_m^2 + k_m \mu_m 
\end{equation*}

Similarly, for $\sigma^2_p$, we consider $\frac{d}{dt} \sum_{(m,n)}n^2 p_{(m,n)}$ and ultimately find that

\begin{equation*}
\frac{d\sigma_p^2}{dt} = \frac{d}{dt}(\sum_{(m,n)}n^2p_{(m,n)}) - 2\mu_p \frac{d\mu_p}{dt}\\
= k_2 \mu_m + k_p\mu_p - 2k_p\sigma^2_p + 2k_2\sigma_{mp}
\end{equation*}

where

\begin{equation*}
\sigma_{mp} = \sum_{(m,n)}mnp_{(m,n)} - \mu_m \mu_p.
\end{equation*}

We see then that we will also need to find the equation for $\sigma_{mp}$.  Thus, we consider $\frac{d}{dt} \sum_{(m,n)}mn p_{(m,n)}$.  Following the usual procedure, we find 

\begin{equation*}
\frac{d}{dt}\sigma_{mp} = \frac{d}{dt}(\sum_{(m,n)}mnp_{(m,n)}) - \frac{d\mu_p}{dt}\mu_m - \mu_p \frac{d\mu_m}{dt}\\
= - k_m\sigma_{mp}+k_2\sigma^2_m - k_p \sigma_{mp}
\end{equation*}

At steady-state where the time derivatives are zero, 
\begin{equation*}
\tilde{k}_1 =\mu_m, \\
\mu_m=\frac{k_p \mu_p}{k_2}, \\
\sigma_m^2 = \mu_m, \\
\sigma_{mp} = \frac{k_2}{k_m + k_p} \sigma^2_m, \\
\sigma_p^2 = \mu_p + \frac{k_2}{k_p} \sigma_{mp}.
\end{equation*}

Based on these equations, we see that
\begin{equation*}
\frac{\sigma_m^2}{\mu_m^2} = \frac{1}{\mu_m} \\ 
\frac{\sigma_p^2}{\mu_p^2} = \frac{1}{\mu_p} + \frac{k_2}{k_p} \frac{k_2}{k_m+k_p} \frac{\sigma_m^2}{\mu^2_p} \\
= \frac{1}{\mu_p} + \frac{\gamma}{1+\gamma} \frac{\sigma_m^2}{\mu_m^2}
\end{equation*}
where $\gamma = \frac{k_p}{k_m}$.

\item If $\gamma \gg 1$ i.e. the protein decays much faster than the mRNA, then the variability in the protein concentration will be inherited from the mRNA values.

If $\gamma \ll 1$, then we have $\frac{\sigma_p^2}{\mu_p^2} \approx \frac{1}{\mu_p}$ which is independent of mRNA variability. 


\end{enumerate}
